## Helpful functions

```python
def select_weeks(df, weeks):
    """
    Filter the DataFrame to include only rows where LAST_X_WEEKS equals the specified number of weeks.
    
    Parameters:
        df (DataFrame): The DataFrame containing the data.
        weeks (int): The number of weeks to filter by.
    
    Returns:
        DataFrame: The filtered DataFrame containing only rows with LAST_X_WEEKS equal to the specified weeks.
    """
    return df[df['LAST_X_WEEKS'] == weeks]
```

```python
def print_badge_associations(column_name):
    """
    Print the names of badges associated with the given column name.
    
    Parameters:
        column_name (str): The column name for which to print badge associations.
    """
    badge_associations = {
        "EVENTS_CREATED": "📆 Event Planner",
        "EVENT_PARTICIPANTS": "📆 Event Planner",
        "POSTS_CREATED": "🗣️ Conversation Starter",
        "REPLIES_RECEIVED": "🗣️ Conversation Starter",
        "ITEMS_GIFTED": "🎁 Philanthropist",
        "THANKYOUS_RECEIVED": "🫱🏻 Helping Hand",
        "PLACES_RECOMMENDED": "🧳 Local Guide"
    }
    
    if column_name in badge_associations:
        print(f"{badge_associations[column_name].upper()}")
    else:
        print(f"No badge association found for '{column_name}'.")
```

```python
def visualize_columns(df, column_names):
    """
    Visualize specified columns from the DataFrame using seaborn countplot.
    
    Parameters:
        df (DataFrame): The DataFrame containing the data.
        column_names (list): A list of column names to visualize.
    """
       
    for column_name in column_names:
        # Print associated badge name
        print_badge_associations(column_name)
        
        plt.figure(figsize=(15, 19))
        ax = sns.countplot(x=df[column_name])
#         plt.title(f"Countplot of {column_name}")
        
        # Set integer x-axis tick labels
        plt.xticks(rotation=45)  # Rotate labels for better visibility
        plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: int(x)))
        
        # Add count labels to the bars
        for p in plt.gca().patches:
            
            plt.gca().annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
        
        sns.despine(bottom=True, right=True, left=True)
        ax.set(yticklabels=[])  
        
        plt.tick_params(axis='both', which='both', bottom=False, left=False)
        plt.xlim(1, 25)
        
        plt.ylim(0, df[column_name].value_counts()[1.0])
        plt.show()
        
        ax.set_ylabel(None)    
```

```python
# List of columns to visualize
columns_to_visualize = ['POSTS_CREATED', 'REPLIES_RECEIVED', 'THANKYOUS_RECEIVED',
                        'EVENTS_CREATED', 'EVENT_PARTICIPANTS', 'ITEMS_GIFTED',
                        'PLACES_RECOMMENDED']
```

```python
def statistics(df, column_names, weeks):
    df = select_weeks(df, weeks)
    
    stats = {}

    for column_name in column_names:
        stats[column_name] = {
            'Max': df[column_name].max(),
            'Min': df[column_name].min(),
            'Mean': df[column_name].mean(),
            'Median': df[column_name].median(),
            'Std Dev': df[column_name].std(),
            'Count': df[column_name].count()
        }

    return stats
```

###  Statistics per category per time frame

```python
def analyze_columns(df, columns):
    for column in columns:
        print(f'👉🏻 {column}')
        print()
        print(f'Mean: {round(df[column].mean(),2)}')
        print(f'Max: {df[column].max()}')
        print()
        print('🔎 Examining unique values')
        print(f'Median: {np.median(df[column].unique())}')
        print(f'Mean: {round(np.mean(df[column].unique()),2)}')
        
        # Spread of the middle portion of the dataset
        q1 = np.percentile(df[column].unique(), 25)
        q3 = np.percentile(df[column].unique(), 75)
        iqr = q3 - q1
        print(f'q1: {q1}')
        print(f'q3: {q3}')
        print(f'Interquartile Range (IQR): {iqr}')
        
        print('-'*25)
        print()
```

```python
# 12 months
analyze_columns(df_stats_12, columns_to_visualize)
```

### Aggregated time frame

```python
# REMOVING TOP USER: Four dataframes containing user activity for a specific time period
df_4weeks = df_stats[df_stats['LAST_X_WEEKS'] == 4].set_index('USER_ID')
df_6weeks = df_stats[df_stats['LAST_X_WEEKS'] == 6].set_index('USER_ID')
df_8weeks = df_stats[df_stats['LAST_X_WEEKS'] == 8].set_index('USER_ID')
df_12weeks = df_stats[df_stats['LAST_X_WEEKS'] == 12].set_index('USER_ID')
```

```python
# One month ago
one_month = df_4weeks

# 6 weeks ago
one_month_and_half = df_6weeks - df_4weeks

# Two months ago
two_months = df_8weeks - (one_month_and_half + one_month)

# Three months ago
three_months = df_12weeks - (two_months + one_month_and_half + one_month)
```

```python
def calculate_stats(df, columns):
    stats = {}  # Dictionary to store statistics for each column
    for column in columns:
        unique_values = df[column].unique()
        median_unique = np.median(unique_values)
        max_value = np.max(unique_values)
        stats[column] = (median_unique, max_value)  # Store stats in dictionary
    return stats

stats = calculate_stats(one_month, columns_to_visualize)
stats2 = calculate_stats(two_months, columns_to_visualize)
stats3 = calculate_stats(three_months, columns_to_visualize)
```

```python
stats2
```

```python
def plot_stats(stats_list):
    fig, axs = plt.subplots(2, figsize=(10, 10))
    
    # Plot median values
    for stats in stats_list:
        for column, (median, _) in stats.items():
            axs[0].plot(column, median, marker='o', linestyle='-')

    axs[0].set_title('Median')
    axs[0].tick_params(axis='x', rotation=45)
    axs[0].spines['top'].set_visible(False)
    axs[0].spines['bottom'].set_visible(False)
    axs[0].spines['right'].set_visible(False)
    axs[0].spines['left'].set_visible(False)
    
    # Plot maximum values
    for stats in stats_list:
        for column, (_, max_val) in stats.items():
            axs[1].plot(column, max_val, marker='x')
    
    axs[1].set_title('Maximum')
    axs[1].tick_params(axis='x', rotation=45)
    axs[1].spines['top'].set_visible(False)
    axs[1].spines['bottom'].set_visible(False)
    axs[1].spines['right'].set_visible(False)
    axs[1].spines['left'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Example usage:
plot_stats([stats, stats2, stats3])
```
